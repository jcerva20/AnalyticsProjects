```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
```

At first tried using Beautifulsoup by itself, however it was not able to pull the data. 

The solution was to use the selenium library with the beautifulsoup library by using the webdrive function to act like a normal user requesting the html info. 

```python
product_titles = []
product_prices = []
product_rating =[]
product_total_ratings = []
product_link = []
product_specs = []

# creates website url, that is going to be scraped
def get_url(search_term): 
    sample_url = "https://www.amazon.com/s?k={}&ref=nb_sb_noss_1"
    search_term = search_term.replace(" ","+")

    search_term += "&page={}"
    
    return sample_url.format(search_term)

# Scraping the data from the url that is generated
def scrape_data(all_listings):
    for listing in all_listings: #loop that iterates through all the product listings
        sponsored = listing.find("span",{"class":"s-label-popover-hover","class":"a-color-base"}).text
        if sponsored != "Sponsored": # Filters out all the sponsored listings 
            title = listing.h2.a.text # gets the product name 
            price = listing.find("span",{"class":"a-offscreen"}).text # gets the product price
            rating = listing.i.text # gets the product rating
            T_ratings = listing.find("span",{"class":"a-size-base s-underline-text"}).text # gets the reviews of the products
            try:
                # gets the specs for monitors, such as display size, resolution, response time, refresh rate, etc. 
                specs = str(listing.find("div",{"class":"a-section a-spacing-none a-spacing-top-small s-product-grid-adjustment"}).find_all("span",{"class":"a-text-bold"})).replace('<span class="a-text-bold">',"").replace("</span>","")
            except:
                specs = None
            url = "https://amazon.com"+listing.find("h2",{"class":"a-size-mini a-spacing-none a-color-base s-line-clamp-2"}).a.get("href")
            product_titles.append(title)
            product_prices.append(price)
            product_rating.append(rating) 
            product_total_ratings.append(T_ratings)
            product_link.append(url)
            product_specs.append(specs)

def webscrape_main(search_term):

    url = get_url(search_term)

    pages = int(input("Number of pages?:"))

    # creates a page iteration, used to scrape multiple pages of a search term 
    for page in range(1,pages + 1):
        driver.get(url.format(page)) # searches the url using the opened browser
        soup = BeautifulSoup(driver.page_source,"lxml") # gets the html from the web page
        all_listings = soup.find_all("div",{"data-component-type":"s-search-result"})
        scrape_data(all_listings)
```

Run the main program:

```python
driver = webdriver.Chrome(executable_path=r"C:\SeleniumDrivers\chromedriver.exe") #opens browser
webscrape_main("ultrawide monitor") # runs all the functions
```

## Cleaning the data 

```python
pd.set_option("display.max_colwidth",-1) #Displays the full text of a pandas DF without any truncation 
```

```python
amazon_df = pd.DataFrame({"Product_Name":product_titles,"Price":product_prices,"Product Specs":product_specs,"Rating":product_rating,"Customer Reviews":product_total_ratings,"Product Link":product_link})
amazon_df.head()
```

```python
amazon_df[["Display_Size","Refresh_Rate","Resolution","Response_Time"]] = amazon_df["Product Specs"].str.split(",",expand=True)
amazon_df.head()
```

Rearranging the columns:

```python
amazon_df.drop(columns=["Product Specs"],inplace=True)
amazon_df = amazon_df.reindex(columns=["Product_Name","Price","Display_Size","Refresh_Rate","Resolution","Response_Time","Rating","Customer Reviews","Product Link"])
amazon_df.head(2)
```

```python
amazon_df["Display_Size"] = amazon_df["Display_Size"].apply(lambda x: x.replace("[",""))
amazon_df.head(2)
```

```python
amazon_df["Response_Time"] = amazon_df["Response_Time"].apply(lambda x: x.replace("]",""))
amazon_df.head(2)
```
